## Comune di Firenze
### Aggregazione indicatori su efficienza e facilità di utilizzo del servizio 

http://annuario.comune.fi.it/dataset?sort=score+desc%2C+metadata_modified+desc&q=&organization=annuario-statistico&tags=servizi+online

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob, os
```

```python
eff_files = glob.glob('efficienza/*.csv')
print (eff_files)
```

```python
fac_files = glob.glob('facilita/*.csv')
print (fac_files)
```

```python
dic_efficienza = {
                 "14-agevolazione-tariffaria-efficienza.csv":"Agevolazione tariffaria",
                 "50-sportello-edilizia-efficienza.csv":"Sportello edilizia",
                 "68-contributo-scuola-efficienza.csv":"Contributo scuola",
                 "23-iscrizione-asili-nido-efficienza.csv":"Iscrizione asili nido",
                 "41-dichiarazione-tari-efficienza.csv":"Dichiarazione-tari",
                 "32-iscrizione-scuola-infanzia-efficienza.csv":"Iscrizione scuola infanzia",
                 "59-appuntamento-edi-urb-efficienza.csv":"Appuntamento edi urb"
                 }

dic_facilita = {
                 "26-iscrizione-scuola-infanzia-facilita-nel-trovare-il-servizio.csv":"Iscrizione scuola infanzia",
                 "44-sportello-edilizia-facilita-nel-trovare-il-servizio.csv":"Sportello edilizia",
                 "62-contributo-scuola-facilita-nel-trovare-il-servizio.csv":"Contributo scuola",
                 "35-dichiarazione-tari-facilita-nel-trovare-il-servizio.csv":"Dichiarazione-tari",
                 "53-appuntamento-edi-urb-facilita-nel-trovare-il-servizio.csv":"Appuntamento edi urb",
                 "17-iscrizione-asili-nido-facilita-nel-trovare-il-servizio.csv":"Iscrizione asili nido",
                 "8-agevolazione-tariffaria-facilita-nel-trovare-il-servizio.csv":"Agevolazione tariffaria"
                 }
```

```python
df_eff = pd.concat([pd.read_csv(fp, sep=",").assign(New=os.path.basename(fp)) for fp in eff_files])
df_eff.rename(columns={'New': 'Sorgente'}, inplace=True)
```

```python
# Facilità servizi - encoding differente
```

```python
df_fac = pd.concat([pd.read_csv(fp, sep=",",  encoding="ISO-8859-1").assign(New=os.path.basename(fp)) for fp in fac_files])
df_fac.rename(columns={'New': 'Sorgente'}, inplace=True)
```

```python
df_eff
```

```python
df_fac
```

```python
# rinominare i campi in sorgente con un dictionary
```

```python
df_eff.replace({"Sorgente": dic_efficienza}, inplace=True)
```

```python
df_fac.replace({"Sorgente": dic_facilita}, inplace=True)
```

## Salviamo i due file

```python
df_eff.to_csv("efficienza.csv", index=None)
```

```python
df_fac.to_csv("facilita.csv", index=None)
```
